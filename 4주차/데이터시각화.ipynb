#### 시각화 옵션- 제목/범례/색상/축이름

```python
# 제목추가
import matplotlib.pyplot as plt
plt.title('Line')
plt.plot([1,2,3], [1,2,3], marker='s') # 마커가 사각형 모양으로 보여짐
plt.show()
```

```python
# 범례추가
plt.title('Legend')
plt.plot([1,2,3,4], label='asc') # 증가 
plt.plot([4,3,2,1], label='desc') # 감소
plt.legend()
plt.show()
```

```python
# 색상
plt.plot([1,2,3], [1,2,3], c='r')
plt.show()
```

```python
# 축이름
plt.plot([1,2,3], [1,2,3])
plt.xlabel("X")
plt.ylabel("Y Line")
plt.show()
```

```python
# 선모양 변경
plt.plot([1,2,3,4], color='r', linestyle='--', label='dashed') # 빨간색 
plt.plot([4,3,2,1], color='g', linestyle=':', label='dotted') # 감소
plt.legend()
plt.show()
```

```python
# 그림 범위
plt.title("x, y 레인지") # 처음은 한글이 꺠지는데 밑에 저걸 실행시키면 한글로 잘 나온다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r") # 색, 두꺼움, 마커는 o 
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns

#한글타이틀
plt.rc('font', family='Malgun Gothic') 

#데이터 준비 
iris = sns.load_dataset("iris")    # 붓꽃 데이터
# titanic = sns.load_dataset("titanic")   # 타이타닉호 데이터
# tips = sns.load_dataset("tips")    # 팁 데이터
# lights = sns.load_dataset("flights")    # 여객운송 데이터
# 머신러닝에 많이 사용하는 데이터
iris
```

```python
#데이터 준비 
tips = sns.load_dataset("tips")    # 팁 데이터

#기본 배경 설정
sns.set_palette("pastel")

#시각화
sns.stripplot(x="day", y="total_bill", data=tips)

# 개인화
plt.title('팁데이터')
plt.ylabel("요금")
plt.xlabel("요일") # 금요일에 돈을 많이 낸다고 알 수 있다. 요일은 범주형 요굼은 연속형으로 혼합해서 쓸 수 있다.
```

#### 막대그래프

```python
#평균치 빠르게 집계해서 시각화하기
sns.barplot(x= "sex",y= "tip",data= tips)
tips['tip'].groupby(tips['sex'],observed=False).mean()
```

```python

# estimator(함수를 지정을 안하면 그냥 나오는데, 데이터 길이를 함수로 해서 길이만큼 나오면 남자가 지불한 게 숫자로는 더 많고 대수적으로 따지면 남자가 더 많다.)를 지정하면 된다. (여기에 들어올 값으로는 list에 해당하는 함수를 사용할 수 있다.)
sns.barplot(x= "sex",y= "tip",estimator=len,data= tips)
```

```python
#여러 열에서 집단 묶어서 세부 집단 시각화 하기 (집단을 나누게 된다.)
# hue 파라미터 추가
sns.barplot(x= "sex",y= "tip",hue="day", data= tips)
```

```python
# 표준편차 오차막대
sns.barplot(x= "sex",y= "tip",hue="day",data= tips,ci="sd") #ci를 다른 걸로 바꿔라. 오래됐으니 sd로 바꾼다. -> 표준편차와 분산을 예측 가능하다. 눈으로
```

#### 연속형

```python
#커널밀도함수
#sns.kdeplot(데이터프레임.변수)
sns.kdeplot(tips.tip) # 2달러 팁이 더 많다.
```

```python
#sns.distplot(데이터프레임.변수)
sns.displot(tips.tip) # 2달러 변수쪽에 숫자가 제일 많아 여기에서 팁을 제일 많이 주는 경우가 많았다.
```

#### 이변수

```python
#연속형/연속형
```

```python
#jointplot(x="x_name", y="y_name", data=dataframe, kind='scatter')
sns.jointplot(x="total_bill", y="tip", data=tips) # 연속형 2개인 경우 상관계수를 찍어볼 수 있다. -> 상관이 많이 비례로 커진다라는 걸 알 수 있다. 
```

```python
#커널밀도함수
sns.jointplot(x="total_bill", y="tip", data=tips,kind="kde") # kde로 선으로 표현하겠다.
```

```python
#범주형/범주형 두개 다 범주형 일때...
```

```python
tips_var= tips.pivot_table(index="smoker", columns="sex", aggfunc="size") # 사이즈를 가지고 평균을 구함 -> 사이즈는 갯수라고 보면 됨 -> 프린트하면 갯수가 나옴
print(tips_var)
sns.heatmap(tips_var, cmap=sns.light_palette("gray", as_cmap=True), annot=True, fmt="d")
```
